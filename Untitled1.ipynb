```python
import sqlite3
import pandas as pd
conn = sqlite3.connect("database.db")
```

```python
# consulta dos dados no banco de dados
consulta_atividade = """
SELECT
  *
 FROM
 flight_activity fa LEFT JOIN flight_loyalty_history flh ON (fa.loyalty_number = flh.loyalty_number)

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )
# visualiza os dados
df_atividade.head()
```

```python
# Numero de dados faltantes
df_atividade.isna().sum()
# selecionando somente as colunas numéricas
colunas=["year", "month", "flights_booked", "flights_with_companions", "total_flights", "distance", "points_accumulated","salary","clv", "loyalty_card"]#o loyalty card serve pra
df_colunas_numericas = df_atividade.loc[:, colunas]
# removendo linhas com alguma coluna vazia.
df_dados_completos = df_colunas_numericas.dropna()#comando do pandas que apaga
df_dados_completos.isna().sum()
```

```python
df_dados_completos.shape[0]#mostra a linha
```

```python
df_atividade.info()
```

```python
from sklearn import tree as tr
X_atributo = df_dados_completos.drop( columns = "loyalty_card")
y = df_dados_completos.loc[:, "loyalty_card"]
# Comando: Defina o algoritmo Decision Tree
modelo = tr.DecisionTreeClassifier( max_depth=7 )#This means that the tree can make decisions based on at most two features or conditions.
# Comando: Treine o algoritmo Decision Tree
modelo_treinado = modelo.fit( X_atributo, y )#x atributo e y rotulo
```

```python
tr.plot_tree(modelo_treinado, filled=True );
```

```python
y.head()
```

```python
X_novo = X_atributo.sample()
previsao = modelo_treinado.predict_proba(X_novo)
print("AURORA: {:.1f}% - NOVA {:.1f}% - START: {:.1f}%".format(100*previsao[0][0],100*previsao[0][1],100*previsao[0][2]))
```
